```python
import matplotlib.pyplot as plt

# Data for the models and their accuracies
models = [
    "Word2Vec + KNN", 
    "BERT + BertForSequenceClassification", 
    "BERT + Logistic Regression (SK-Learn)", 
    "BERT + Logistic Regression (PyTorch)", 
    "BERT + MLPClassifier", 
    "Sentence Transformers + Logistic Regression (PyTorch)", 
    "Sentence Transformers + MLPClassifier", 
    "Sentence Transformers + Linear SVM"
]
accuracies = [75.3, 93.9, 90.9, 90.9, 90.9, 90.9, 86.4, 79.5]

# Creating the plot
plt.figure(figsize=(12, 6))
plt.bar(models, accuracies, color='skyblue')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy of Different Models for Binary Classification Task')
plt.xticks(rotation=45, ha='right')
plt.ylim(70, 100)  # Adjusting y-axis limits for better visibility
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the models and their accuracies for the 4-class classification task
models_4_classes = [
    "Word2Vec + KNN", 
    "BERT + BertForSequenceClassification", 
    "BERT + Logistic Regression (SK-Learn)", 
    "BERT + Logistic Regression (PyTorch)", 
    "BERT + MLPClassifier", 
    "Sentence Transformers + Logistic Regression (PyTorch)", 
    "Sentence Transformers + MLPClassifier", 
    "Sentence Transformers + Linear SVM"
]
accuracies_4_classes = [48.8, 72.7, 56.8, 61.4, 56.8, 56.8, 59.1, 59.1]

# Creating the plot for the 4-class classification task
plt.figure(figsize=(12, 6))
plt.bar(models_4_classes, accuracies_4_classes, color='lightcoral')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy of Different Models for 4-Class Classification Task')
plt.xticks(rotation=45, ha='right')
plt.ylim(40, 80)  # Setting y-axis limits to better visualize the differences
plt.tight_layout()
plt.show()
```
